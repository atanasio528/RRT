```python
import numpy as np
import matplotlib.pyplot as plt

# Three state markov chain and Ross recovery
states = ['Bull', 'Stagnant', 'Bear']
s2i = {s: i for i, s in enumerate(states)}
i2s = {i: s for i, s in enumerate(states)}

# Transition matrix under Q
trans_mtx = np.array([
    [0.7, 0.2, 0.1], # Bull
    [0.2, 0.6, 0.2], # Stagnant
    [0.1, 0.3, 0.6], # Bear
])

init_state = 'Stagnant'
init_idx = s2i[init_state]

def simulate_mc(steps, trans_mtx, init_idx):
    states_seq = [init_idx]
    for _ in range(steps):
        current_idx = states_seq[-1]
        next_idx = np.random.choice(
            a = len(states),
            p = trans_mtx[current_idx]
        )
        states_seq.append(next_idx)
    return states_seq

steps = 1000
simulation_idx = simulate_mc(steps, trans_mtx, init_idx)
simulated_states = [i2s[i] for i in simulation_idx]

plt.figure(figsize=(12, 2))
plt.plot(simulated_states, marker='o', linestyle='-', color='black')
plt.xticks(rotation=90)
plt.title("Simulated Market Regime using Time-Homogeneous Markov Chain")
plt.ylabel("State")
plt.grid(True)
plt.tight_layout()
plt.show()
```
